```python
import sys
import platform
print(f'Platform: {sys.platform}')
plat = sys.platform
```

```python
#If os is windows and gives the user info if they are on the most recent version
if plat == "win32":
    current_ver = platform.version()
    latest_ver = "10.0.26200"
    if current_ver < latest_ver:
        print("System not up to date.")
    elif current_ver == latest_ver:
        print("System is up to date!")
    else:
        print("System exceeds common avaliable version.")
```

```python
#If user uses the linux os, this gives them info if they are using the most recent verison
if plat == "linux":
    current_ver = platform.version()
    latest_ver = "6.19.12"
    if current_ver < latest_ver:
        print("System not up to date.")
    elif current_ver == latest_ver:
        print("System is up to date!")
    else:
        print("System exceeds common avaliable version.")
    print("Linux")  
```

```python
if plat == "darwin":
    print("Mac")
```
